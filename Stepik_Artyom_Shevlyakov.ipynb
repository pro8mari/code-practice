```python
#Создайте две новых таблицы: army_persons - это все студенты, которых могут призвать в армию из-за отчисления, и free_persons - это те, кого не призовут.
#Подсчитайте среднюю длину волос у студентов таблиц army_persons и free_persons .
army_persons = df[df['Army'] == 'могут призвать']
free_persons = df[df['Army'] == 'не призовут (по разным причинам)']
army_persons['Hair length'].mean(), free_persons['Hair length'].mean()
```

```python
#Найдите всех студентов, к которых 0 друзей в социальных сетях. Сколько среди них женщин и мужчин?
df[df['Friend number'] == 0].value_counts('Sex')
```

```python
#Создайте таблицу dor_lovers - это все студенты, которые любят собак.
#В ответ запишите долю (процент) из них, которые живут на первом этаже.
dog_lovers = df[df['Animal'] == 'Собак'].loc[:,['Animal','Floor number']]
len(dog_lovers[dog_lovers['Floor number'] == 1])/len(dog_lovers)
```

```python
#Получить две таблицы: single_child - это все студенты у которых нет ни брата ни сестры
#not_single_child - это все студенты у которых есть брат или сестра
single_child = df[df['Brother-sister'] == 'нет ни брата ни сестры']['Social network duration min']
not_single_child = df[df['Brother-sister'] != 'нет ни брата ни сестры']['Social network duration min']
#Найдите медиану признака 'Social network duration min' (сколько времени студент проводит в соцсетях) отдельно для студентов таблицы  single_child  и таблицы not_single_child .
single_child.median() - not_single_child.median()
```

```python
#Создайте три таблицы: KFC_persons - студенты, которые ходят в KFC
#Burger_persons - студенты, которые ходят в Burger king
#Mac_persons - студенты, которые ходят в Макдональдс (aka "Вкусно и точка")
#Для каждой категории студентов подсчитайте средний вес студентов.
KFC_persons = df[df['Fastfood'] == 'KFC']['Weight']
Burger_persons = df[df['Fastfood'] == 'Бургер кинг']['Weight']
Mac_persons = df[(df['Fastfood'] != 'Бургер кинг') & (df['Fastfood'] != 'KFC')]['Weight']
Mac_persons.mean() , Burger_persons.mean(), KFC_persons.mean()
```
